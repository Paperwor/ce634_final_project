```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('C:\\files\\Documents\\polyu\\courses\\urban_big_data\\final_project\\train.csv')

district_crime_counts = data['PdDistrict'].value_counts().sort_values(ascending=False)
total_crimes = district_crime_counts.sum()
crime_proportion = (district_crime_counts / total_crimes) * 100  # Calculate proportion as percentage

fig, ax1 = plt.subplots(figsize=(12, 8))

bars = district_crime_counts.plot(kind='bar', color='skyblue', ax=ax1)
ax1.set_title('Crime Count and Proportion by Police District')
ax1.set_xlabel('Police District')
ax1.set_ylabel('Number of Crimes', color='b')
ax1.tick_params(axis='y', labelcolor='b')
ax1.set_xticklabels(district_crime_counts.index, rotation=45)
ax1.grid(True)

for i, value in enumerate(district_crime_counts):
    ax1.text(i, value + 5, str(value), ha='center', va='bottom', color='b')


ax2 = ax1.twinx()  
line = crime_proportion.plot(kind='line', marker='o', color='red', ax=ax2)
ax2.set_ylabel('Proportion of Total Crimes (%)', color='r')
ax2.tick_params(axis='y', labelcolor='r')

plt.show()
```
