```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


data = pd.read_csv('C:\\files\\Documents\\polyu\\courses\\urban_big_data\\final_project\\train.csv')


print(data.head(100))
```

```python
data['Dates'] = pd.to_datetime(data['Dates'])

data['Year'] = data['Dates'].dt.year

data['Year'] = data['Dates'].dt.year

yearly_counts = data['Year'].value_counts().sort_index()

plt.figure(figsize=(12, 8))
bars = plt.bar(yearly_counts.index.astype(str), yearly_counts.values, color='skyblue')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), va='bottom', ha='center')  

plt.title('Number of Crimes Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()  
plt.show()
```

```python
filtered_yearly_counts = yearly_counts.drop(2015)

average_crimes_per_year = filtered_yearly_counts.mean()
print(f"Average number of crimes per year excluding 2015: {average_crimes_per_year:.2f}")

variance_crimes_per_year = filtered_yearly_counts.var()
print(f"Variance of crimes per year excluding 2015: {variance_crimes_per_year:.2f}")
```

```python
data['Month'] = data['Dates'].dt.month

monthly_counts = data['Month'].value_counts().sort_index()

month_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

plt.figure(figsize=(10, 6))
bars = plt.bar(month_labels, monthly_counts, color='skyblue')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), va='bottom', ha='center')  

plt.title('Number of Crimes Per Month')
plt.xlabel('Month')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
data['Hour'] = data['Dates'].dt.hour

hourly_counts = data['Hour'].value_counts().sort_index()

plt.figure(figsize=(16, 9))
bars = plt.bar(hourly_counts.index, hourly_counts, color='skyblue')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), va='bottom', ha='center')  

plt.title('Number of Crimes Per Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Crimes')
plt.xticks(hourly_counts.index, [f'{i}:00' for i in hourly_counts.index])  
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```
